```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
n_samples = 100
n_points = 10
```

```python
plt.style.use('ggplot')
```

```python
func_dict = {1: plt.plot, 2: plt.scatter, 3: plt.bar, 4: plt.pie, 5: plt.imshow}
labels_dict = {1: 'Line', 2: 'Scatter', 3: 'Bar', 4: 'Pie', 5: 'HeatMap'}
```

```python
labels = []
names = []
```

```python
def get_x_data(i):
    x_data = np.random.randn(n_points)
    if i == 1 or i == 3:
        return np.arange(len(x_data)), x_data
    elif i == 2:
        return x_data, np.random.randn(n_points)
    elif i == 4:
        return x_data, None
    elif i == 5:
        return x_data[:9].reshape(3, 3), None
```

```python
plt.ioff()
```

```python
my_dpi = 300
```

```python
for i in np.arange(n_samples):
    plt.figure(figsize=(300//my_dpi, 300//my_dpi), dpi=my_dpi)
    func = np.random.choice([1, 2, 3, 4, 5])
    x_data, y_data = get_x_data(func)
    if y_data is None:
        func_dict[func](x_data)
    else:
        func_dict[func](x_data, y_data)
    plt.savefig('../Images/'+str(i)+'.png')
    plt.close()
    labels.append(func)
    names.append(labels_dict[func])
```

```python
with open('../Labels/labels.txt', 'wb') as fp:
    fp.write('\n'.join(str(label) for label in labels))
```

```python
with open('../Labels/names.txt', 'wb') as fp:
    fp.write('\n'.join(str(label) for label in labels))
```
